```python
datas = np.arange(1,88,1)
```

```python
datas
```

```python
plt.plot(datas, Moscow)
```

```python
Y = Moscow
X = np.arange(len(Moscow)) # transformando variáveis contínuas em um conjunto linear
n = np.size(Moscow)
```

**calculando o coeficiente angular da reta**

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
```

**calculando o coeficiente linear da reta**

```python
b = np.mean(Y) - a*np.mean(X)
```

**calculando a reta**

```python
y = a*X+b
```

**calculando o RMSE para transformar a distância euclidiana em unidades reais do negócio**

```python
norma = np.linalg.norm(Moscow-y)
n = len(Moscow)
rmse = norma / np.sqrt(n)
rmse
```

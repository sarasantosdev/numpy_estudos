### Entendendo o comportamento do dados

No gráfico abaixo podemos ver que os preços sobem e descem em diversos momentos, mas não podemos dizer que o preço da maçã sobe no decorrer do tempo, para resolver isso vamos criar uma reta que nos informa a taxa de crescimento. Para calcular essa reta faremos a seguinte equação:

**y = ax + b**, onde y equivalente ao preço das maçãs, a é o coeficiente angular que desloca a reta e o b é o coeficiente linear onde a reta corta o eixo y, e o x corresponde ao mês

```python
plt.plot(datas, Moscow)
```

```python
x = datas
y = 2*x + 80
```

```python
plt.plot(datas, Moscow)
plt.plot(x, y)
```

```python
y = 0.52*x + 80
```

```python
plt.plot(datas, Moscow)
plt.plot(x, y)
```

```python
np.linalg.norm(Moscow-y)
```

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
```

```python
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X+b
y.size
```

```python
np.linalg.norm(Moscow-y)
```

```python
# verificando se o erro é alto
norma = np.linalg.norm(Moscow[-11:]-y)
n = 11
rmse = norma / np.sqrt(n)
rmse
```

### Gerando números aleatórios

```python
np.random.seed(84)
coef_angulares = np.random.uniform(low=0.10, high=0.90, size=100)
```

```python
norma = np.array([])
for i in range(100):
    norma = np.append(norma,np.linalg.norm(Moscow-(coef_angulares[i]*X+b)))
```

```python
norma
```

```python
coef_angulares
```

```python
np.random.seed(16)
np.random.uniform(low=0.10, high=0.90, size=100)
```

### Salvando os dados

```python
dados = np.column_stack([norma, coef_angulares])
```

```python
dados.shape
```

```python
np.savetxt('dados.csv', dados, delimiter=',')
```
